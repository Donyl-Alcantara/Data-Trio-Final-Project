```python
import yfinance as yf
import pandas as pd
```

```python
sgb = yf.download("SGB=F", start="2015-01-01", 
                  auto_adjust=False, multi_level_index=False)
```

```python
sgb
```

```python
sgb.isnull().sum()
```

```python
sgb = sgb[["Close"]].copy()
sgb.columns = ["mops_gasoil"]
sgb = sgb.reset_index()
sgb.columns = ["date", "mops_gasoil"]
sgb = sgb.set_index("date").sort_index()
```

```python
# weekly aggregation (just close + mean since high/low/close are basically same)
weekly_sgb = sgb.resample("W-FRI").last()
weekly_sgb.columns = ["mops_gasoil"]
weekly_sgb["mops_gasoil_mean"] = sgb.resample("W-FRI").mean()

# Anchor to Monday
weekly_sgb.index = weekly_sgb.index - pd.Timedelta(days=4)
weekly_sgb.index.name = "week_start"
```

```python
weekly_sgb.to_parquet("data/semi-cleaned/mops_gasoil_weekly.parquet")
```
