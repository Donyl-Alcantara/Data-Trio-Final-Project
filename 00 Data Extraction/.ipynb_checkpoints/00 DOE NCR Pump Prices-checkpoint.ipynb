```python
def download_doe_pdfs(output_folder='data/raw/doe_pdfs'):
    """Scrapes the DOE portal for retail pump price PDFs."""
    target_url = os.getenv('DOE_TARGET_URL')
    session_cookie = os.getenv('DOE_SESSION_COOKIE')

    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) Chrome/120.0.0.0 Safari/537.36',
        'Cookie': session_cookie
    }

    print("Authenticating with DOE portal...")
    response = requests.get(target_url, headers=headers)

    if "Log in" in response.text or response.status_code != 200:
        print("Recheck Cookie Validity. Authentication failed.")
        return

    soup = BeautifulSoup(response.text, 'html.parser')
    pdf_links = []

    for link in soup.find_all('a', href=True):
        href = link['href']
        if 'documents/d/' in href.lower() or '-pdf' in href.lower() or href.lower().endswith('.pdf'):
            full_url = urljoin(target_url, href)
            pdf_links.append(full_url)

    pdf_links = list(set(pdf_links))
    print(f"Found {len(pdf_links)} protected PDF(s). Downloading...")

    for pdf_url in pdf_links:
        file_name = pdf_url.split('/')[-1]

        if not file_name.endswith('.pdf'):
            file_name = file_name + '.pdf'

        file_path = os.path.join(output_folder, file_name)

        pdf_response = requests.get(pdf_url, headers=headers)
        pdf_response.raise_for_status()

        with open(file_path, 'wb') as file:
            file.write(pdf_response.content)
        print(f"Saved: {file_name}")

    print("DOE PDF extraction complete.")
```

```python
download_doe_pdfs()
```

```python
download_folder = 'raw/doe_pdfs'
os.makedirs(download_folder, exist_ok=True)

base_website_url = "https://prod-cms.doe.gov.ph"

# --- NEW: Recursive function to hunt for raw URLs hidden in the JSON ---
def extract_raw_urls(data):
    """Deep-searches the JSON dictionary for any string that looks like a document URL."""
    urls = []
    if isinstance(data, dict):
        for key, value in data.items():
            # If we find a string with '/documents/' that ISN'T a block of HTML, it's a raw link!
            if isinstance(value, str) and '/documents/' in value and '<' not in value:
                urls.append(value)
            # Otherwise, keep digging deeper
            elif isinstance(value, (dict, list)):
                urls.extend(extract_raw_urls(value))
    elif isinstance(data, list):
        for item in data:
            urls.extend(extract_raw_urls(item))
    return urls


print(f"\nStarting downloads for {len(all_data)} items...")

for item in all_data:
    # We use a 'set' to collect links so we don't accidentally download the same file twice
    links_to_download = set()
    
    # 1. Hunt for Attachments: Search the raw JSON structure
    raw_urls = extract_raw_urls(item)
    for url in raw_urls:
        links_to_download.add(url)
        
    # 2. Hunt for Inline Links: Search the HTML fields (like we did before)
    for field in item.get('contentFields', []):
        if 'contentFieldValue' in field and 'data' in field['contentFieldValue']:
            html_content = str(field['contentFieldValue']['data'])
            soup = BeautifulSoup(html_content, 'html.parser')
            for a_tag in soup.find_all('a'):
                href = a_tag.get('href', '')
                if href:
                    links_to_download.add(href)
                    
    # 3. Process and download all the links we found
    for href in links_to_download:
        
        # Check if it's a PDF file
        if href.endswith('-pdf') or href.endswith('.pdf'):
            
            # Convert relative URLs to absolute URLs
            full_download_url = urljoin(base_website_url, href)
            
            # Clean up the filename
            raw_filename = href.split('/')[-1]
            if raw_filename.endswith('-pdf'):
                clean_filename = raw_filename[:-4] + ".pdf"
            else:
                clean_filename = raw_filename
                
            filepath = os.path.join(download_folder, clean_filename)
            
            if os.path.exists(filepath):
                print(f"Already exists: {clean_filename}. Skipping.")
                continue
                
            print(f"Downloading: {clean_filename}...")
            
            try:
                # Assuming 'headers' and 'cookies' from your scraping loop are still defined above this
                pdf_response = requests.get(full_download_url, headers=headers, cookies=cookies)
                
                if pdf_response.status_code == 200:
                    with open(filepath, 'wb') as f:
                        f.write(pdf_response.content)
                else:
                    print(f"  -> Error: Status {pdf_response.status_code}")
                    
                time.sleep(0.5)
                
            except Exception as e:
                print(f"  -> Failed: {e}")

print("\nAll downloads completed!")
```
