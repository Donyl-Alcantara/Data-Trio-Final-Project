```python
import pandas as pd
```

```python
gpr = pd.read_csv("../data/semi-cleaned/ai_gpr_data_daily.csv", parse_dates=['Date'])
```

```python
gpr = gpr[gpr['Date'] > '2015']
```

```python
gpr
```

```python
gpr = gpr.set_index('Date')

# relevant columns
gpr = gpr[['GPR_AI', 'GPR_OIL', 'GPR_OIL_MiddleEast', 'GPR_OIL_Russia', 'GPR_NONOIL']]

# filter to study period
gpr = gpr.loc['2018-01-01':]
```

```python
# Use mean of the week, not last — GPR is a flow variable, not a price
gpr_weekly = gpr.resample('W-TUE').mean()
```

```python
gpr_weekly
```

```python
gpr_weekly.to_parquet('../data/semi-cleaned/gpr_weekly.parquet')
```
