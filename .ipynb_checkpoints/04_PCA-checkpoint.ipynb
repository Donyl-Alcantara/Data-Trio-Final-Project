# Notebook 04: PCA / Dimensionality Reduction
**Philippine NCR Pump Price Forecasting**

This notebook applies PCA exactly as your professor's PCA notebook demonstrates, adapted to the pump price feature set.

### Why PCA here?
After Feature Engineering (Notebook 03), we have ~40+ columns. Many are correlated (e.g., `brent_mean`, `brent_lag2`, `brent_lag3` are 0.95+ correlated). PCA:
1. Eliminates multicollinearity that destabilizes OLS coefficients
2. Reduces dimensionality for cleaner model inputs
3. **Required by DMW rubric** — retain 95% cumulative variance

**Research basis**: Lunor et al. (2023) §2.2: *'datasets were processed through PCA to reduce multicollinearity and increase model interpretability... retained 95% cumulative variability.'*

### Two PCA applications
1. **Lag-feature block PCA** — compress the lag columns (most correlated block)
2. **Cross-product price PCA** — decompose the 4 pump price series into latent components (for EDA insight, not modelling input)

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

import os
os.makedirs('../outputs', exist_ok=True)

print('Libraries loaded.')
```

```python
df = pd.read_parquet('../data/final/feature_matrix.parquet')

PRODUCTS = ['diesel', 'ron91', 'ron95', 'ron97']
available_products = [p for p in PRODUCTS if p in df.columns]

print('Feature matrix shape:', df.shape)
df.head(3)
```

## Part 1: Lag Feature Block PCA

Following your professor's PCA notebook methodology:
1. Identify the lag-feature block (most collinear columns)
2. Standardize (mean=0, std=1) — critical before PCA
3. Fit PCA, plot cumulative explained variance
4. Retain components at 95% threshold (Lunor et al. standard)
5. Interpret what each PC represents

**Important**: In modelling, the scaler and PCA are fit ONLY on the training fold, then applied to test. This cell fits on ALL data for interpretability only.

```python
# Identify lag columns
lag_cols = [c for c in df.columns if '_lag' in c and c not in available_products]
print(f'Lag columns ({len(lag_cols)} total):')
print(lag_cols)
```

```python
# Step 1: Select lag block, drop rows with any NaN in this block
X_lag = df[lag_cols].dropna()
print(f'Lag block shape: {X_lag.shape}')

# Step 2: Standardize
scaler = StandardScaler()
X_lag_scaled = scaler.fit_transform(X_lag)

# Check standardization
print(f'Mean after scaling (should be ~0): {X_lag_scaled.mean():.6f}')
print(f'Std  after scaling (should be ~1): {X_lag_scaled.std():.6f}')
```

```python
# Step 3: Fit full PCA to see all variance explained
pca_full = PCA()
pca_full.fit(X_lag_scaled)

# Step 4: Plot cumulative explained variance (professor's plot_cum_exp_var style)
exp_var = pca_full.explained_variance_ratio_
cum_var = np.cumsum(exp_var)

# Find threshold at 95%
n_95 = np.argmax(cum_var >= 0.95) + 1

fig, ax = plt.subplots(figsize=(9, 6))
ax.plot(range(1, len(cum_var)+1), cum_var, 'o-', lw=2.5, markersize=5)
ax.axvline(n_95, linestyle='--', lw=2, color='tab:orange')
ax.axhline(0.95, linestyle=':', lw=1.5, color='gray')
ax.text(n_95 + 0.3, 0.05, f'PCs retained: {n_95}', color='tab:orange', weight='bold', fontsize=12)
ax.set_xlim([0, len(cum_var)+1])
ax.set_ylim([0, 1.05])
ax.set_xlabel('Number of Components', fontsize=12)
ax.set_ylabel('Cumulative Explained Variance', fontsize=12)
ax.set_title('Cumulative Explained Variance vs Number of PCs\n(Lag Feature Block)', fontsize=13, weight='bold')
ax.grid(True, alpha=0.3)
ax.spines[['top','right']].set_visible(False)
plt.tight_layout()
plt.savefig('../outputs/04a_pca_scree_lag_block.png', dpi=150, bbox_inches='tight')
plt.show()

print(f'PCs needed to explain 95% variance: {n_95} (out of {len(lag_cols)} original lag features)')
print(f'Dimensionality reduction: {len(lag_cols)} → {n_95} columns')
print()
print('Variance explained per PC:')
for i, v in enumerate(exp_var[:n_95]):
    print(f'  PC{i+1}: {v*100:.2f}% (cumulative: {cum_var[i]*100:.2f}%)')
```

```python
# Step 5: Fit PCA at 95% threshold
pca_lag = PCA(n_components=0.95)
X_pca_lag = pca_lag.fit_transform(X_lag_scaled)

print(f'PCA components retained: {pca_lag.n_components_}')
print(f'Variance explained: {pca_lag.explained_variance_ratio_.sum()*100:.2f}%')

# Store as DataFrame
pc_lag_cols = [f'lag_PC{i+1}' for i in range(pca_lag.n_components_)]
df_pca_lag = pd.DataFrame(X_pca_lag, index=X_lag.index, columns=pc_lag_cols)

df_pca_lag.head(3)
```

```python
# Step 6: Interpret PCs — what original features drive each PC?
W = pd.DataFrame(
    pca_lag.components_,
    columns=lag_cols,
    index=pc_lag_cols
)

# Show loadings for PC1 and PC2 (professor's W matrix style)
print('PC1 top loadings (largest |magnitude|):')
pc1_sorted = W.loc['lag_PC1'].abs().sort_values(ascending=False)
print(pc1_sorted.head(8).round(3))

print('\nPC2 top loadings:')
pc2_sorted = W.loc['lag_PC2'].abs().sort_values(ascending=False)
print(pc2_sorted.head(8).round(3))

# Heatmap of loadings
fig, ax = plt.subplots(figsize=(14, 0.5*min(len(lag_cols), 8) + 2))
top_n = min(pca_lag.n_components_, 5)
W_top = W.iloc[:top_n]
sns.heatmap(W_top.T, cmap='RdBu_r', center=0, ax=ax,
            annot=True, fmt='.2f', annot_kws={'size': 7},
            linewidths=0.3)
ax.set_title('PCA Loadings — Lag Feature Block (top 5 PCs)', fontsize=11, weight='bold')
ax.set_xlabel('Principal Component')
ax.set_ylabel('Original Feature')
plt.tight_layout()
plt.savefig('../outputs/04b_pca_loadings_lag.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Part 2: Cross-Product Price PCA

Following your professor's PCA analysis: PCA on the 4 pump price time series tells us how much of the variation in all prices is explained by a common 'fuel price level' factor (PC1) vs a 'diesel vs gasoline spread' factor (PC2). This is the Lunor et al. (2023) approach to understanding cross-product dynamics.

```python
price_df = df[available_products].dropna()

# Standardize
scaler_price = StandardScaler()
X_price_scaled = scaler_price.fit_transform(price_df)

# Full PCA
pca_price = PCA()
pca_price.fit(X_price_scaled)

exp_var_price = pca_price.explained_variance_ratio_
cum_var_price = np.cumsum(exp_var_price)

print('Pump Price PCA — Variance Explained:')
for i, (ev, cv) in enumerate(zip(exp_var_price, cum_var_price)):
    print(f'  PC{i+1}: {ev*100:.2f}% (cumulative: {cv*100:.2f}%)')

# W matrix (loadings)
W_price = pd.DataFrame(
    pca_price.components_,
    columns=available_products,
    index=[f'PC{i+1}' for i in range(len(available_products))]
)
print('\nPC Loadings:')
print(W_price.round(3))
```

```python
# Scree plot for pump prices
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Left: cumulative variance
axes[0].bar(range(1, len(exp_var_price)+1), exp_var_price*100, color='steelblue', alpha=0.8)
axes[0].plot(range(1, len(cum_var_price)+1), cum_var_price*100, 'o-', color='tomato', lw=2)
axes[0].axhline(95, linestyle='--', color='gray', lw=1)
axes[0].set_xlabel('Principal Component')
axes[0].set_ylabel('Variance Explained (%)')
axes[0].set_title('PCA Scree Plot — Pump Prices', weight='bold')
axes[0].grid(True, alpha=0.3)

# Right: loadings heatmap
sns.heatmap(W_price.T, cmap='RdBu_r', center=0, ax=axes[1],
            annot=True, fmt='.2f', annot_kws={'size': 9}, linewidths=0.5)
axes[1].set_title('PC Loadings — Pump Price Products', weight='bold')

plt.tight_layout()
plt.savefig('../outputs/04c_pca_pump_prices.png', dpi=150, bbox_inches='tight')
plt.show()

print('\nInterpretation:')
print('  PC1: if loadings are similar across products → "overall fuel price level" factor')
print('  PC2: if diesel loads differently from RON products → "diesel vs gasoline spread"')
print('  This justifies per-product modelling (Lunor et al. found diesel and diesel plus')
print('  had different significant features from gasoline products)')
```

## Part 3: Build Final Model-Ready Datasets

We create two versions of X:
- **X_raw**: original features (for tree models — don't need PCA)
- **X_pca**: PCA-compressed lag block + non-lag features (for linear models)

Saved to `data/final/` for use in the modelling notebooks.

```python
# Non-lag features (go in both X_raw and X_pca as-is)
non_lag_features = [
    c for c in df.columns
    if c not in available_products
    and '_lag' not in c
    and c != 'month'   # raw month number, we use dummies
]
non_lag_features = [c for c in non_lag_features if c in df.columns]

# Align index to rows that have lag data
common_idx = X_lag.index.intersection(df.index)
df_aligned  = df.loc[common_idx]
pca_aligned = df_pca_lag.loc[common_idx]

# X_raw: all features, no PCA
all_feature_cols = lag_cols + non_lag_features
all_feature_cols = [c for c in all_feature_cols if c in df_aligned.columns]
X_raw = df_aligned[all_feature_cols].copy()

# X_pca: PCA-compressed lags + non-lag features
X_pca = pd.concat([pca_aligned, df_aligned[non_lag_features]], axis=1)

# Save both
X_raw.to_parquet('../data/final/X_raw.parquet')
X_pca.to_parquet('../data/final/X_pca.parquet')

# Save target variables
y = df_aligned[available_products]
y.to_parquet('../data/final/y_targets.parquet')

print('Saved to data/final/:')
print(f'  X_raw.parquet    — shape: {X_raw.shape}')
print(f'  X_pca.parquet    — shape: {X_pca.shape}')
print(f'  y_targets.parquet — shape: {y.shape}')
print()
print('X_pca dimensionality reduction:')
print(f'  Lag features: {len(lag_cols)} → {pca_lag.n_components_} PCA components')
print(f'  Non-lag features: {len(non_lag_features)}')
print(f'  Total X_pca columns: {X_pca.shape[1]}')
```

## Summary

| Dataset | Columns | Use |
|---|---|---|
| `X_raw` | All lag + non-lag features (no PCA) | Random Forest, XGBoost (scale-invariant, don't need PCA) |
| `X_pca` | PCA-compressed lags + non-lag features | Ridge, Lasso, SVR (benefit from PCA + StandardScaler) |
| `y_targets` | diesel, ron91, ron95, ron97 | All regression models |

In the modelling notebooks, StandardScaler and PCA will be fit **inside each training fold** using TimeSeriesSplit to prevent data leakage.
