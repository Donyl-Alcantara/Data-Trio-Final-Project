# Notebook 06: Modelling — All Products, All Models
**Philippine NCR Pump Price Forecasting**

This notebook trains, evaluates, and compares ALL models across ALL pump price products:
- **Products**: Diesel, Diesel Plus, Kerosene, RON 91, RON 95, RON 97, RON 100
- **Forecast horizons**: h=1, h=2, h=4 weeks ahead
- **Model tiers** (methodology doc §3.4):

| Tier | Models |
|---|---|
| Tier 0 — Baselines | Persistence (naive), Seasonal Naive, MOPS-formula baseline |
| Tier 1 — Linear | OLS, Ridge, Lasso, ElasticNet |
| Tier 2 — Nonlinear | SVR (RBF), Random Forest, Gradient Boosting, XGBoost |

**Evaluation metrics** (all reported — methodology doc §3.6):
- MAE (₱/L) — most interpretable
- RMSE (₱/L) — penalizes large errors
- MAPE (%) — comparable to published benchmarks (Lu et al. 2021: 0.131%–19.2%)
- MASE — beats persistence? MASE < 1 means yes
- Directional Accuracy (%) — predicted correct direction of price change?

**Split**: Walk-forward TimeSeriesSplit (5 folds, gap=1) — never random KFold.
Results saved to `outputs/model_results.csv` for the final report.

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import os

from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet, RidgeCV, LassoCV, ElasticNetCV
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import TimeSeriesSplit, RandomizedSearchCV
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error, mean_squared_error

try:
    from xgboost import XGBRegressor
    HAS_XGB = True
except ImportError:
    HAS_XGB = False
    print('XGBoost not installed. Run: pip install xgboost')

os.makedirs('../outputs', exist_ok=True)
os.makedirs('../data/final', exist_ok=True)
print('Libraries loaded. XGBoost available:', HAS_XGB)
```

## Load Data

```python
X_raw = pd.read_parquet('../data/final/X_raw.parquet')
y_all = pd.read_parquet('../data/final/y_targets.parquet')
X_pca = pd.read_parquet('../data/final/X_pca.parquet')

# All products present in the target file
ALL_PRODUCTS = [c for c in y_all.columns if c in [
    'diesel', 'diesel_plus', 'kerosene', 'ron91', 'ron95', 'ron97', 'ron100'
]]
print('Products to model:', ALL_PRODUCTS)
print('X_raw shape:', X_raw.shape)
print('Date range:', X_raw.index.min(), '->', X_raw.index.max())
y_all[ALL_PRODUCTS].tail(3)
```

## Helper Functions — Metrics and Evaluation

```python
def mae(y_true, y_pred):
    return mean_absolute_error(y_true, y_pred)

def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

def mape(y_true, y_pred):
    """Mean Absolute Percentage Error — Lu et al. (2021) benchmark metric."""
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    mask = y_true != 0
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100

def mase(y_true, y_pred, y_naive):
    """
    Mean Absolute Scaled Error.
    MASE < 1: beats persistence. MASE > 1: worse than persistence.
    y_naive: persistence forecast (y_t = y_{t-1})
    """
    mae_model = mae(y_true, y_pred)
    mae_naive = mae(y_true, y_naive)
    return mae_model / (mae_naive + 1e-10)

def directional_accuracy(y_true, y_pred, y_prev):
    """% of weeks where predicted direction of price change matches actual."""
    actual_dir = np.sign(np.array(y_true) - np.array(y_prev))
    pred_dir   = np.sign(np.array(y_pred) - np.array(y_prev))
    return np.mean(actual_dir == pred_dir) * 100

def evaluate(y_true, y_pred, y_prev, y_naive, model_name, product, horizon):
    """Return a dict of all metrics for one model/product/horizon combination."""
    return {
        'model':    model_name,
        'product':  product,
        'horizon':  horizon,
        'n_test':   len(y_true),
        'MAE':      round(mae(y_true, y_pred), 4),
        'RMSE':     round(rmse(y_true, y_pred), 4),
        'MAPE':     round(mape(y_true, y_pred), 4),
        'MASE':     round(mase(y_true, y_pred, y_naive), 4),
        'DirAcc':   round(directional_accuracy(y_true, y_pred, y_prev), 2),
    }

print('Metric functions defined.')
```

## Tier 0 Baselines

**Per methodology doc §3.4**: You MUST beat these to claim ML adds value.
- **Persistence**: ŷ_{t+h} = y_t (last known price)
- **Seasonal Naive**: ŷ_{t+h} = y_{t+h-52} (same week last year)
- **MOPS-formula**: ŷ = α + β·(MOPS_{t-1} × USD_PHP_{t-1}) — encodes DOE pricing mechanism

```python
def persistence_forecast(y_series, horizon):
    """Persistence: predict using last observed value."""
    return y_series.shift(horizon)

def seasonal_naive_forecast(y_series, horizon, period=52):
    """Seasonal naive: same week last year."""
    return y_series.shift(period)

print('Baseline forecasters defined.')
```

## Main Modelling Loop

For each product × horizon, we:
1. Build X and y (shift y by horizon for direct multi-step)
2. Walk-forward TimeSeriesSplit (5 folds)
3. Fit each model on training fold, predict on test fold
4. Compute all metrics
5. Store results

**Note on scaling**: StandardScaler is fit inside each fold (only on train) to prevent leakage.

```python
# Define model registry
def get_models(has_xgb=True):
    models = {
        # Tier 1 — Linear (use Pipeline with StandardScaler)
        'OLS':         Pipeline([('scaler', StandardScaler()), ('model', LinearRegression())]),
        'Ridge':       Pipeline([('scaler', StandardScaler()), ('model', Ridge(alpha=1.0))]),
        'Lasso':       Pipeline([('scaler', StandardScaler()), ('model', Lasso(alpha=0.01, max_iter=100000))]),
        'ElasticNet':  Pipeline([('scaler', StandardScaler()), ('model', ElasticNet(alpha=0.01, l1_ratio=0.5, max_iter=100000))]),
        # Tier 2 — Nonlinear
        'SVR':         Pipeline([('scaler', StandardScaler()), ('model', SVR(kernel='rbf', C=10, gamma='scale', epsilon=0.1))]),
        'RF':          RandomForestRegressor(n_estimators=100, max_depth=6, max_features=0.7, random_state=42, n_jobs=-1),
        'GBM':         GradientBoostingRegressor(n_estimators=100, max_depth=3, learning_rate=0.05, random_state=42),
    }
    if has_xgb:
        models['XGBoost'] = XGBRegressor(n_estimators=100, max_depth=3, learning_rate=0.05,
                                          subsample=0.8, colsample_bytree=0.8,
                                          random_state=42, verbosity=0)
    return models

HORIZONS = [1, 2, 4]  # weeks ahead
N_SPLITS  = 5

print('Model registry defined.')
print('Models:', list(get_models(HAS_XGB).keys()))
print('Horizons:', HORIZONS)
print('Products:', ALL_PRODUCTS)
```

```python
# MAIN LOOP — runs for all products × horizons × models
# Expected runtime: ~5-15 min depending on machine

all_results = []
all_predictions = {}  # store predictions for later plotting

tscv = TimeSeriesSplit(n_splits=N_SPLITS, gap=1)

for product in ALL_PRODUCTS:
    if product not in y_all.columns:
        print(f'  Skipping {product} — not in y_all')
        continue

    print(f'\n{'='*60}')
    print(f'PRODUCT: {product.upper()}')
    print(f'{'='*60}')

    for horizon in HORIZONS:
        print(f'  Horizon h={horizon}...')

        # Align X and y, shift y forward by horizon (direct multi-step)
        y_series = y_all[product]
        common_idx = X_raw.index.intersection(y_series.dropna().index)
        X_h = X_raw.loc[common_idx].fillna(0)
        y_h = y_series.loc[common_idx].shift(-horizon)  # target is h weeks ahead

        # Drop rows where target is NaN (end of series)
        valid = y_h.notna()
        X_h, y_h = X_h[valid], y_h[valid]

        # --- TIER 0 BASELINES (on full test portion = last fold) ---
        last_train_end = int(len(X_h) * 0.8)
        y_test_idx = y_h.index[last_train_end:]

        # Raw y (not shifted) for baseline computations
        y_raw_aligned = y_series.loc[X_h.index]

        # Persistence
        persist_pred = y_raw_aligned.shift(horizon).loc[y_test_idx]
        y_true_test  = y_h.loc[y_test_idx]
        y_prev_test  = y_raw_aligned.shift(1).loc[y_test_idx]  # y_{t-1} for directional acc

        valid_test = y_true_test.notna() & persist_pred.notna() & y_prev_test.notna()
        if valid_test.sum() < 5:
            continue

        y_t  = y_true_test[valid_test].values
        y_p  = y_prev_test[valid_test].values
        y_naive_vals = persist_pred[valid_test].values

        all_results.append(evaluate(y_t, y_naive_vals, y_p, y_naive_vals, 'Persistence', product, horizon))

        # Seasonal naive
        sn_pred = y_raw_aligned.shift(52).loc[y_test_idx]
        sn_valid = y_true_test.notna() & sn_pred.notna() & y_prev_test.notna()
        if sn_valid.sum() >= 5:
            y_t2 = y_true_test[sn_valid].values
            all_results.append(evaluate(y_t2, sn_pred[sn_valid].values, y_prev_test[sn_valid].values,
                                        persist_pred[sn_valid].values, 'SeasonalNaive', product, horizon))

        # MOPS-formula baseline (if feature exists)
        if 'mops_in_php_gasoil' in X_h.columns:
            mops_x = X_h[['mops_in_php_gasoil']]
            mops_train = mops_x.iloc[:last_train_end]
            mops_test  = mops_x.iloc[last_train_end:]
            y_mops_train = y_h.iloc[:last_train_end]
            lr_mops = LinearRegression().fit(mops_train, y_mops_train)
            mops_pred = lr_mops.predict(mops_test.loc[y_test_idx])
            mops_valid = y_true_test.notna() & y_prev_test.notna()
            if mops_valid.sum() >= 5:
                y_t3 = y_true_test[mops_valid].values
                all_results.append(evaluate(y_t3, mops_pred[:mops_valid.sum()], y_prev_test[mops_valid].values,
                                            persist_pred[mops_valid].values, 'MOPSFormula', product, horizon))

        # --- TIER 1 & 2 ML MODELS (walk-forward CV) ---
        models = get_models(HAS_XGB)

        for model_name, model in models.items():
            fold_preds = []
            fold_trues = []
            fold_prevs = []
            fold_naiverefs = []

            for train_idx, test_idx in tscv.split(X_h):
                X_tr = X_h.iloc[train_idx]
                X_te = X_h.iloc[test_idx]
                y_tr = y_h.iloc[train_idx]
                y_te = y_h.iloc[test_idx]

                # y_prev for this fold (for directional accuracy)
                y_prev_fold = y_raw_aligned.shift(1).iloc[test_idx]
                y_naive_fold = y_raw_aligned.shift(horizon).iloc[test_idx]

                # Drop NaN rows in this fold
                valid_tr = y_tr.notna()
                valid_te = y_te.notna() & y_prev_fold.notna() & y_naive_fold.notna()

                if valid_tr.sum() < 20 or valid_te.sum() < 3:
                    continue

                try:
                    model.fit(X_tr[valid_tr], y_tr[valid_tr])
                    preds = model.predict(X_te[valid_te])
                    fold_preds.extend(preds)
                    fold_trues.extend(y_te[valid_te].values)
                    fold_prevs.extend(y_prev_fold[valid_te].values)
                    fold_naiverefs.extend(y_naive_fold[valid_te].values)
                except Exception as e:
                    pass

            if len(fold_trues) >= 5:
                result = evaluate(
                    np.array(fold_trues), np.array(fold_preds),
                    np.array(fold_prevs), np.array(fold_naiverefs),
                    model_name, product, horizon
                )
                all_results.append(result)

                # Store last fold predictions for plotting
                key = (product, model_name, horizon)
                all_predictions[key] = {
                    'true': np.array(fold_trues[-len(y_test_idx):]),
                    'pred': np.array(fold_preds[-len(y_test_idx):])
                }

print('\n✅ Modelling complete!')
print(f'Total results: {len(all_results)}')
```

## Multi-Horizon Comparison — Best Model per Product

```python
# For each product, show MAPE across horizons for all models
# Exclude baselines for clarity
ml_models = [m for m in results_df['model'].unique() if m not in ['Persistence','SeasonalNaive','MOPSFormula']]
baselines  = ['Persistence', 'SeasonalNaive', 'MOPSFormula']

n_products = len(ALL_PRODUCTS)
n_horizons = len(HORIZONS)

fig, axes = plt.subplots(n_products, 1, figsize=(12, 4*n_products), sharex=False)
if n_products == 1:
    axes = [axes]

for ax, product in zip(axes, ALL_PRODUCTS):
    prod_df = results_df[results_df['product'] == product]
    # group by model and horizon, average MAPE
    pivot = prod_df.pivot_table(index='model', columns='horizon', values='MAPE')
    
    # Plot ML models
    for model in ml_models:
        if model in pivot.index:
            ax.plot(HORIZONS, [pivot.loc[model, h] if h in pivot.columns else np.nan for h in HORIZONS],
                    'o-', linewidth=1.5, markersize=6, label=model)
    # Plot baselines as dashed
    for b in baselines:
        if b in pivot.index:
            ax.plot(HORIZONS, [pivot.loc[b, h] if h in pivot.columns else np.nan for h in HORIZONS],
                    '--', linewidth=1.2, alpha=0.7, label=b)

    ax.set_title(f'{product.upper()} — MAPE by Horizon', fontsize=10, weight='bold')
    ax.set_xlabel('Forecast Horizon (weeks ahead)')
    ax.set_ylabel('MAPE (%)')
    ax.set_xticks(HORIZONS)
    ax.legend(fontsize=7, ncol=3)
    ax.grid(True, alpha=0.3)

plt.suptitle('MAPE by Forecast Horizon — All Products', fontsize=13, weight='bold', y=1.01)
plt.tight_layout()
plt.savefig('../outputs/06d_mape_by_horizon_all_products.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Forecast vs Actual Plots — Best Model, h=1, All Products

```python
# Find best model per product at h=1 (lowest MAPE among ML models)
h1_ml = results_df[(results_df['horizon'] == 1) & results_df['model'].isin(ml_models)]
best_per_product = h1_ml.loc[h1_ml.groupby('product')['MAPE'].idxmin()][['product','model','MAPE']]
print('Best model per product (h=1, by MAPE):')
print(best_per_product.to_string(index=False))
```

```python
# Re-train best model on 80% and plot predictions on held-out 20%
fig, axes = plt.subplots(len(ALL_PRODUCTS), 1, figsize=(14, 4*len(ALL_PRODUCTS)), sharex=False)
if len(ALL_PRODUCTS) == 1:
    axes = [axes]

for ax, product in zip(axes, ALL_PRODUCTS):
    if product not in y_all.columns:
        continue

    # Find best model for this product
    bpm = best_per_product[best_per_product['product'] == product]
    if bpm.empty:
        continue
    best_model_name = bpm.iloc[0]['model']
    best_mape = bpm.iloc[0]['MAPE']

    # Build data
    y_series = y_all[product]
    common_idx = X_raw.index.intersection(y_series.dropna().index)
    X_h = X_raw.loc[common_idx].fillna(0)
    y_h = y_series.loc[common_idx].shift(-1)  # h=1
    valid = y_h.notna()
    X_h, y_h = X_h[valid], y_h[valid]

    split = int(len(X_h) * 0.8)
    X_tr, X_te = X_h.iloc[:split], X_h.iloc[split:]
    y_tr, y_te = y_h.iloc[:split], y_h.iloc[split:]

    models = get_models(HAS_XGB)
    if best_model_name not in models:
        continue
    model = models[best_model_name]
    try:
        model.fit(X_tr, y_tr)
        y_pred = model.predict(X_te)
    except:
        continue

    # Plot
    ax.plot(y_te.index, y_te.values, linewidth=1.5, label='Actual', color='black')
    ax.plot(y_te.index, y_pred, linewidth=1.5, label=f'{best_model_name} (MAPE={best_mape:.2f}%)',
            color='tomato', linestyle='--')
    # Persistence as baseline
    y_persist = y_series.shift(1).loc[y_te.index]
    ax.plot(y_te.index, y_persist.values, linewidth=0.8, label='Persistence',
            color='gray', linestyle=':', alpha=0.7)

    ax.set_title(f'{product.upper()} — Best Model: {best_model_name} | MAPE={best_mape:.2f}%', fontsize=10, weight='bold')
    ax.set_ylabel('₱/L')
    ax.legend(fontsize=8)
    ax.grid(True, alpha=0.3)
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

plt.suptitle('Forecast vs Actual — Best Model per Product (h=1, held-out 20%)', fontsize=12, weight='bold', y=1.01)
plt.tight_layout()
plt.savefig('../outputs/06e_forecast_vs_actual_all_products.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Residual Analysis

```python
# Residuals for diesel at h=1 with best model
TARGET = 'diesel' if 'diesel' in ALL_PRODUCTS else ALL_PRODUCTS[0]

y_series = y_all[TARGET]
common_idx = X_raw.index.intersection(y_series.dropna().index)
X_h = X_raw.loc[common_idx].fillna(0)
y_h = y_series.loc[common_idx].shift(-1)
valid = y_h.notna()
X_h, y_h = X_h[valid], y_h[valid]
split = int(len(X_h) * 0.8)

bpm = best_per_product[best_per_product['product'] == TARGET]
best_model_name = bpm.iloc[0]['model'] if not bpm.empty else 'Ridge'
models = get_models(HAS_XGB)
model = models[best_model_name]
model.fit(X_h.iloc[:split], y_h.iloc[:split])
y_pred_test = model.predict(X_h.iloc[split:])
residuals = y_h.iloc[split:].values - y_pred_test

fig, axes = plt.subplots(1, 3, figsize=(14, 4))

# Residuals over time
axes[0].plot(y_h.index[split:], residuals, linewidth=0.8, color='steelblue')
axes[0].axhline(0, color='red', linestyle='--', linewidth=1)
axes[0].set_title('Residuals Over Time')
axes[0].set_ylabel('Residual (₱/L)')
axes[0].grid(True, alpha=0.3)

# Residual histogram
axes[1].hist(residuals, bins=40, color='steelblue', edgecolor='white', linewidth=0.3)
axes[1].axvline(0, color='red', linestyle='--')
axes[1].set_title('Residual Distribution')
axes[1].set_xlabel('Residual (₱/L)')
axes[1].grid(True, alpha=0.3)

# Predicted vs Actual
axes[2].scatter(y_h.iloc[split:].values, y_pred_test, alpha=0.4, s=15, color='steelblue')
lims = [min(y_h.iloc[split:].min(), y_pred_test.min()),
        max(y_h.iloc[split:].max(), y_pred_test.max())]
axes[2].plot(lims, lims, 'r--', linewidth=1)
axes[2].set_xlabel('Actual (₱/L)')
axes[2].set_ylabel('Predicted (₱/L)')
axes[2].set_title('Predicted vs Actual')
axes[2].grid(True, alpha=0.3)

plt.suptitle(f'Residual Analysis — {TARGET.upper()}, {best_model_name}, h=1', fontsize=11, weight='bold')
plt.tight_layout()
plt.savefig('../outputs/06f_residual_analysis.png', dpi=150, bbox_inches='tight')
plt.show()

print(f'Residual mean:  {residuals.mean():.4f} ₱/L (should be ~0)')
print(f'Residual std:   {residuals.std():.4f} ₱/L')
print(f'Max residual:   {np.abs(residuals).max():.4f} ₱/L')
```

## Benchmark Comparison vs Literature

From methodology doc §3.6 and §10:
- Lu et al. (2021) decade review: MAPE range = **0.131% – 19.2%**
- Lunor et al. (2023) PH pump prices: MAPE = **3.13% – 12.67%**
- Our target: MAE < ₱1.50/L, MAPE < 5% for h=1

```python
print('=' * 70)
print('BENCHMARK COMPARISON vs PUBLISHED LITERATURE')
print('=' * 70)
print()
print('Lu et al. (2021) decade review — MAPE range: 0.131% – 19.2%')
print('Lunor et al. (2023) PH pump prices — MAPE range: 3.13% – 12.67%')
print()

# Our best results at h=1
best_h1 = results_df[(results_df['horizon']==1) & results_df['model'].isin(ml_models)]
best_h1_summary = best_h1.groupby('product')['MAPE'].min().reset_index()
best_h1_summary.columns = ['Product', 'Best MAPE (%)']
best_h1_summary['In Lunor range?'] = best_h1_summary['Best MAPE (%)'].apply(
    lambda x: '✅' if 3.13 <= x <= 12.67 else ('🔻 Better!' if x < 3.13 else '⚠️ Higher')
)
best_h1_summary['In Lu range?'] = best_h1_summary['Best MAPE (%)'].apply(
    lambda x: '✅' if x <= 19.2 else '⚠️ Check'
)
print(best_h1_summary.to_string(index=False))

overall_best = best_h1_summary['Best MAPE (%)'].mean()
print(f'\nMean best MAPE across all products: {overall_best:.2f}%')
print(f'Target from methodology doc: < 5% for h=1 → {"✅ ACHIEVED" if overall_best < 5 else "⚠️ Work on features"}')
```
