# Notebook 05: Feature Importance
**Philippine NCR Pump Price Forecasting**

This notebook answers: **which features actually matter, and how much?**

We use three complementary methods (from methodology doc §3.7):
1. **OLS p-value screening** — Lunor et al. (2023) approach: drop features with p > 0.05
2. **Lasso path** — automatic feature selection via L1 penalty (Notebook 3A style from professor)
3. **Random Forest feature importance** — tree-based, handles nonlinearity (Notebook 4 style)
4. **Permutation importance** — model-agnostic, computed on held-out data

### Critical check
Per our methodology doc §10: *'If your top SHAP/importance feature is anything other than `mops_in_php_lag1` or `brent_lag1 × usd_php_lag1`, something is wrong.'*

We use **Diesel** as the primary product for demonstration. The same analysis runs for all products.

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression, Lasso, LassoCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import TimeSeriesSplit
from sklearn.inspection import permutation_importance

import statsmodels.api as sm
import os
os.makedirs('../outputs', exist_ok=True)

print('Libraries loaded.')
```

```python
X_raw = pd.read_parquet('../data/final/X_raw.parquet')
y_all = pd.read_parquet('../data/final/y_targets.parquet')

# Use diesel as primary product
TARGET = 'diesel'
if TARGET not in y_all.columns:
    TARGET = y_all.columns[0]
    print(f'diesel not found, using: {TARGET}')

# Align and drop rows where target is NaN
common_idx = X_raw.index.intersection(y_all.index)
X = X_raw.loc[common_idx].copy()
y = y_all.loc[common_idx, TARGET].copy()

# Drop rows where either X or y has NaN
valid = X.notna().all(axis=1) & y.notna()
X, y = X[valid], y[valid]

print(f'Target: {TARGET}')
print(f'X shape: {X.shape}')
print(f'y shape: {y.shape}')
print(f'Date range: {X.index.min()} to {X.index.max()}')
```

## Method 1: OLS p-value Screening

**Research basis**: Lunor et al. (2023) §2.2: *'datasets underwent an initial multiple linear regression weight p-value test to determine individual feature significance. A p-value of 0.05 was used.'*

This is the professor's Notebook 3A approach applied to feature selection.

```python
# Standardize before OLS (so coefficients are comparable)
scaler = StandardScaler()
X_scaled = pd.DataFrame(
    scaler.fit_transform(X.fillna(0)),
    index=X.index, columns=X.columns
)

# Fit OLS with statsmodels to get p-values
X_sm = sm.add_constant(X_scaled)
model_ols = sm.OLS(y, X_sm).fit()

# Extract p-values and coefficients
pvals = model_ols.pvalues.drop('const').sort_values()
coefs = model_ols.params.drop('const')

significant = pvals[pvals < 0.05]
print(f'Total features: {len(pvals)}')
print(f'Significant (p<0.05): {len(significant)}')
print(f'Would drop: {len(pvals) - len(significant)}')
print()
print('Top 15 most significant features (lowest p-value):')
print(significant.head(15).round(6))
```

```python
# Visualize: coefficient magnitude of significant features (professor's Notebook 3A bar chart style)
sig_coefs = coefs[significant.index].sort_values()
n_show = min(25, len(sig_coefs))
sig_coefs_show = pd.concat([sig_coefs.head(n_show//2), sig_coefs.tail(n_show//2)]).sort_values()

fig, ax = plt.subplots(figsize=(10, max(6, len(sig_coefs_show)*0.3)))
colors = ['tomato' if v < 0 else 'steelblue' for v in sig_coefs_show]
ax.barh(range(len(sig_coefs_show)), sig_coefs_show.values, color=colors, align='center')
ax.set_yticks(range(len(sig_coefs_show)))
ax.set_yticklabels(sig_coefs_show.index, fontsize=8)
ax.axvline(0, color='black', linewidth=0.8)
ax.set_xlabel('Standardized Coefficient (OLS)')
ax.set_title(f'OLS Feature Coefficients — {TARGET.upper()}\n(significant features, p<0.05)', fontsize=11, weight='bold')
ax.grid(True, alpha=0.3, axis='x')
plt.tight_layout()
plt.savefig('../outputs/05a_ols_coefficients.png', dpi=150, bbox_inches='tight')
plt.show()

print(f'\nOLS R²: {model_ols.rsquared:.4f}')
print(f'Adjusted R²: {model_ols.rsquared_adj:.4f}')
```

## Method 2: Lasso Path — Automatic Feature Selection

**Research basis**: Notebook 3A shows Ridge and Lasso paths across alpha values. Lasso (L1) drives some coefficients to exactly zero — a built-in feature selector. The features that survive at moderate alpha are the robust predictors.

This mirrors the professor's approach: scan alpha range, plot coefficient paths, identify which features survive.

```python
from sklearn.linear_model import Lasso

# Train/test split (temporal — last 20% as test, professor's style but time-aware)
split = int(len(X_scaled) * 0.8)
X_train, X_test = X_scaled.iloc[:split], X_scaled.iloc[split:]
y_train, y_test = y.iloc[:split], y.iloc[split:]

# Scan alpha range (professor's Notebook 3A style)
alpha_lasso = [1e-5, 1e-4, 1e-3, 5e-3, 1e-2, 5e-2, 0.1, 0.2, 0.5, 1.0, 2.0, 5.0, 10.0]
training_r2 = []
test_r2     = []
n_features_used = []

for alpha_run in alpha_lasso:
    lasso = Lasso(alpha=alpha_run, max_iter=100000)
    lasso.fit(X_train, y_train)
    training_r2.append(lasso.score(X_train, y_train))
    test_r2.append(lasso.score(X_test, y_test))
    n_features_used.append(np.sum(lasso.coef_ != 0))

# Plot (professor's errorbar style)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].semilogx(alpha_lasso, training_r2, 'o-', label='Training R²', color='blue', markersize=8)
axes[0].semilogx(alpha_lasso, test_r2, '^-', label='Test R²', color='red', markersize=8)
axes[0].set_xlabel('Alpha (Lasso regularization)', fontsize=11)
axes[0].set_ylabel('R²', fontsize=11)
axes[0].set_title('Lasso — R² vs Alpha', fontsize=11, weight='bold')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

axes[1].semilogx(alpha_lasso, n_features_used, 's-', color='green', markersize=8)
axes[1].set_xlabel('Alpha (Lasso regularization)', fontsize=11)
axes[1].set_ylabel('Number of Features Used', fontsize=11)
axes[1].set_title('Lasso — Feature Count vs Alpha', fontsize=11, weight='bold')
axes[1].grid(True, alpha=0.3)

plt.suptitle(f'Lasso Regularization Path — {TARGET.upper()}', fontsize=12, weight='bold')
plt.tight_layout()
plt.savefig('../outputs/05b_lasso_path.png', dpi=150, bbox_inches='tight')
plt.show()

best_alpha_idx = np.argmax(test_r2)
print(f'Best test R²: {max(test_r2):.4f} at alpha={alpha_lasso[best_alpha_idx]}')
print(f'Features used at best alpha: {n_features_used[best_alpha_idx]}')
```

```python
# Fit Lasso at best alpha and show surviving features
best_alpha = alpha_lasso[best_alpha_idx]
lasso_best = Lasso(alpha=best_alpha, max_iter=100000)
lasso_best.fit(X_train, y_train)

surviving = pd.Series(lasso_best.coef_, index=X.columns)
surviving = surviving[surviving != 0].sort_values()

print(f'\nLasso surviving features at alpha={best_alpha} ({len(surviving)} features):')

fig, ax = plt.subplots(figsize=(10, max(5, len(surviving)*0.35)))
colors = ['tomato' if v < 0 else 'steelblue' for v in surviving.values]
ax.barh(range(len(surviving)), surviving.values, color=colors)
ax.set_yticks(range(len(surviving)))
ax.set_yticklabels(surviving.index, fontsize=8)
ax.axvline(0, color='black', linewidth=0.8)
ax.set_xlabel('Lasso Coefficient')
ax.set_title(f'Lasso Feature Coefficients — {TARGET.upper()} (alpha={best_alpha})', fontsize=11, weight='bold')
ax.grid(True, alpha=0.3, axis='x')
plt.tight_layout()
plt.savefig('../outputs/05c_lasso_coefficients.png', dpi=150, bbox_inches='tight')
plt.show()

weights_norm = surviving / surviving.abs().sum()
print('\nTop 10 features by Lasso weight:')
print(surviving.abs().sort_values(ascending=False).head(10).round(4))
```

## Method 3: Random Forest Feature Importance

**Research basis**: Professor's Notebook 4 §Feature importance in trees. Random Forest importances are more reliable than single decision trees. This is model-agnostic to linear assumptions and captures nonlinear relationships.

```python
from sklearn.ensemble import RandomForestRegressor

# Temporal split (no random shuffle — professor's method applied correctly for time series)
# We use raw X here since trees don't need scaling
X_tr = X.iloc[:split].fillna(0)
X_te = X.iloc[split:].fillna(0)
y_tr = y.iloc[:split]
y_te = y.iloc[split:]

# Random Forest (professor's Notebook 4 style)
rf = RandomForestRegressor(n_estimators=100, max_depth=6, max_features=0.7, random_state=42, n_jobs=-1)
rf.fit(X_tr, y_tr)

print(f'RF Training R²: {rf.score(X_tr, y_tr):.4f}')
print(f'RF Test R²:     {rf.score(X_te, y_te):.4f}')
```

```python
# Feature importance (professor's plot_feature_importances_cancer style)
importances = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)
top_n = 20
top_importances = importances.head(top_n)

fig, ax = plt.subplots(figsize=(10, 7))
ax.barh(range(top_n), top_importances.values[::-1], align='center', color='steelblue')
ax.set_yticks(range(top_n))
ax.set_yticklabels(top_importances.index[::-1], fontsize=9)
ax.set_xlabel('Feature Importance (Gini)', fontsize=11)
ax.set_title(f'Random Forest Feature Importance — {TARGET.upper()} (Top {top_n})', fontsize=11, weight='bold')
ax.grid(True, alpha=0.3, axis='x')
plt.tight_layout()
plt.savefig('../outputs/05d_rf_feature_importance.png', dpi=150, bbox_inches='tight')
plt.show()

print('Top 10 RF features:')
print(top_importances.head(10).round(4))
```

## Method 4: Permutation Importance

**Research basis**: Methodology doc §3.7 — permutation importance is model-agnostic and computed on held-out data to avoid bias. It works by randomly shuffling one feature at a time and measuring the drop in R². Features that cause large drops when shuffled are important.

```python
from sklearn.inspection import permutation_importance

perm_result = permutation_importance(
    rf, X_te.values, y_te.values,
    n_repeats=10, random_state=42, n_jobs=-1
)

perm_imp = pd.DataFrame({
    'importance_mean': perm_result.importances_mean,
    'importance_std':  perm_result.importances_std
}, index=X.columns).sort_values('importance_mean', ascending=False)

top_perm = perm_imp.head(20)

fig, ax = plt.subplots(figsize=(10, 7))
ax.barh(range(len(top_perm)),
        top_perm['importance_mean'].values[::-1],
        xerr=top_perm['importance_std'].values[::-1],
        align='center', color='steelblue', alpha=0.8, capsize=3)
ax.set_yticks(range(len(top_perm)))
ax.set_yticklabels(top_perm.index[::-1], fontsize=9)
ax.axvline(0, color='black', linewidth=0.8)
ax.set_xlabel('Permutation Importance (mean ΔR²)', fontsize=11)
ax.set_title(f'Permutation Importance on Test Set — {TARGET.upper()} (Top 20)', fontsize=11, weight='bold')
ax.grid(True, alpha=0.3, axis='x')
plt.tight_layout()
plt.savefig('../outputs/05e_permutation_importance.png', dpi=150, bbox_inches='tight')
plt.show()

print('Top 10 by permutation importance (test set):')
print(perm_imp.head(10).round(4))
```

## Sanity Check and Summary

```python
print('=' * 70)
print('FEATURE IMPORTANCE SANITY CHECK')
print('=' * 70)
print()

top_ols  = pvals.head(5).index.tolist()
top_rf   = importances.head(5).index.tolist()
top_perm = perm_imp.head(5).index.tolist()

print(f'Top 5 by OLS p-value:           {top_ols}')
print(f'Top 5 by RF importance:         {top_rf}')
print(f'Top 5 by Permutation importance:{top_perm}')
print()

# Check: mops_in_php should dominate
mops_check = any('mops_in_php' in f for f in top_rf[:3])
brent_check = any('brent' in f for f in top_rf[:5])
fx_check = any('usd_php' in f for f in top_rf[:5])

print('Expected dominant features (methodology doc §10):')
print(f'  mops_in_php in top-3 RF: {"✅" if mops_check else "⚠️  CHECK YOUR FEATURES"}')
print(f'  brent in top-5 RF:       {"✅" if brent_check else "⚠️  CHECK YOUR FEATURES"}')
print(f'  usd_php in top-5 RF:     {"✅" if fx_check else "⚠️  CHECK YOUR FEATURES"}')
print()
print('If top features are NOT mops_in_php/brent/usd_php, revisit feature engineering.')
print('The DOE formula guarantees these must dominate.')

# Final feature set recommendation
# Union of significant features across methods
sig_ols  = set(pvals[pvals < 0.05].index)
sig_rf   = set(importances[importances > 0.01].index)
sig_lasso = set(surviving.index)

robust_features = sig_ols.intersection(sig_rf).union(sig_lasso)
print(f'\nRobust feature set (appears significant in ≥2 methods): {len(robust_features)} features')
print('These are the recommended features for the final model in Notebook 06.')
```
