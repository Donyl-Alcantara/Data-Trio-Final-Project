# 03 — Feature Engineering

**Purpose:** Transform the raw master panel into a model-ready feature matrix. Every decision here is grounded in EDA findings from Notebook 02.

**EDA decisions being implemented:**
- WTI dropped (Brent-WTI r=0.991, decided in Cell 6)
- All series are I(1) — lag features built on levels; model learns differencing implicitly
- CCF signal for MOPS→diesel: lag 1 = 0.417, lag 2 = 0.178, decays to noise by lag 4
- Own-price lag-1 autocorrelation ≈ 0.97 across all products — own-price lag is the strongest single predictor
- Rockets and Feathers confirmed: 1.38x asymmetry ratio — asymmetric MOPS features warranted
- 3 nulls in diesel/diesel_plus/kerosene (Jul-Sep 2018), 27 nulls in ron100 (2018-2019)

**What Lunor et al. (2023) did at this stage:**
1. p-value testing per feature per product to drop insignificant features
2. PCA on remaining features retaining 95% cumulative variance (4-5 components per product)
3. Separate feature sets for diesel vs gasoline products

**Our extensions beyond Lunor et al.:**
- Asymmetric MOPS features (motivated by RFH result)
- `mops_in_php` derived feature (encodes the DOE replacement-cost formula directly)
- Crack spread feature
- Event dummies for three structural breaks identified visually in EDA

````python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from scipy import stats
import warnings, os
warnings.filterwarnings('ignore')

df = pd.read_parquet('data/final/master_weekly.parquet')
print('Loaded master:', df.shape)
print('Columns:', df.columns.tolist())
````

## Step 1 — Drop WTI

Decided in EDA Cell 6: Brent-WTI correlation = 0.991. Keeping both causes redundant multicollinearity with no added information. Brent is retained as the primary crude benchmark (globally relevant, daily frequency). Dubai is retained as the Asia-Pacific robustness feature per Lunor et al. (2023).

````python
df = df.drop(columns=['wti_mean'])
print('WTI dropped. Remaining columns:', df.columns.tolist())
````

## Step 2 — Handle Nulls in Target Columns

From EDA Cell 4:
- `diesel`, `diesel_plus`, `kerosene`: 3 nulls each, all in Jul–Sep 2018 (sparse DOE reporting in that period)
- `ron100`: 27 nulls in 2018–2019 (product not widely monitored by DOE in early sample)

**Method:** Forward-fill for both. This is consistent with Lunor et al. (2023) who used constant-within-interval assumption for monthly features interpolated to weekly. The forward-fill assumption is that the price was unchanged in weeks where DOE did not report — defensible given the weekly announcement cadence. Documented as a limitation.

````python
TARGET_COLS = ['diesel', 'diesel_plus', 'kerosene', 'ron100', 'ron91', 'ron95', 'ron97']

print('Nulls BEFORE forward-fill:')
print(df[TARGET_COLS].isnull().sum())

df[TARGET_COLS] = df[TARGET_COLS].ffill()

print('\nNulls AFTER forward-fill:')
print(df[TARGET_COLS].isnull().sum())
print('\nNote: ron100 2018-2019 values are forward-filled estimates.')
print('Models using ron100 should be interpreted with caution for the 2018-2019 period.')
````

## Step 3 — Derived Features

### 3a. `mops_in_php` — The Core Formula Feature
This is the single most important engineered feature. The DOE replacement-cost formula is:
```
PumpPrice_t ≈ MOPS_{t-1} × USD_PHP_{t-1} + taxes + margins
```
Multiplying MOPS (USD/bbl) by the exchange rate (PHP/USD) gives a PHP-denominated input cost. This single feature encodes the mechanism the oil companies themselves use. He (2023) showed that crude oil price alone explains only R²=0.874 of gasoline price variance; adding macro features (which include FX) pushes it to R²=0.947. Our `mops_in_php` combines both into one variable.

### 3b. `crack_spread` — Refining Margin Signal
The crack spread is MOPS minus Brent — the margin between the refined product price and the crude input. When crack spreads widen, refiners are capturing more margin and pump prices tend to rise beyond what crude alone would predict. Lunor et al. (2023) included both crude price and MOPS as separate features; we make the relationship explicit.

### 3c. Asymmetric MOPS Changes — RFH Features
From EDA Cell 21: MOPS-up weeks produced average diesel rise of +0.52, MOPS-down weeks produced average diesel fall of only -0.38 — a 1.38x asymmetry ratio. Following Wen et al. (2025), we split MOPS weekly change into positive and negative components, allowing models to weight them differently.

````python
# Core formula feature
df['mops_in_php'] = df['mops_gasoil_mean'] * df['usd_php_mean']

# Crack spread
df['crack_spread'] = df['mops_gasoil_mean'] - df['brent_mean']

# Asymmetric MOPS changes (RFH features)
mops_chg = df['mops_gasoil_mean'].diff()
df['mops_chg_pos'] = mops_chg.clip(lower=0)   # positive changes only, 0 otherwise
df['mops_chg_neg'] = mops_chg.clip(upper=0)   # negative changes only, 0 otherwise

print('Derived features created:')
print(df[['mops_in_php', 'crack_spread', 'mops_chg_pos', 'mops_chg_neg']].describe().round(3))

# Sanity check: mops_in_php should correlate very strongly with diesel
r = df['mops_in_php'].corr(df['diesel'])
print(f'\nCorrelation mops_in_php vs diesel: {r:.4f}')
print('(Expected > 0.95 — if lower, check units)')
````

## Step 4 — Event Dummies

| Russia-Ukraine war | Feb–Dec 2022 | Invasion Feb 24 2022; price normalization by end 2022 |
| OPEC+/tariff shock | Jan–Apr 2026 | US tariff escalation; visible spike then rapid rollback in EDA plot |

````python
# Brent realized volatility — 4-week rolling std of weekly returns
# Lagged 1 week so it's available at announcement time
brent_return = df['brent_mean'].pct_change()
df['brent_vol_4w'] = brent_return.rolling(4).std().shift(1)

# Same for MOPS — captures product-market specific volatility  
mops_return = df['mops_gasoil_mean'].pct_change()
df['mops_vol_4w'] = mops_return.rolling(4).std().shift(1)
````

## Step 5 — Excise Tax Step Function

TRAIN Law (RA 10963) introduced excise taxes in three tranches. These are step-function changes in the pricing formula — not noise, not a feature to be learned from the data. They must be encoded explicitly. Failure to include them would cause the model to interpret the Jan 2018, Jan 2019, Jan 2020 price jumps as unexplained shocks.

Source: DOE official excise tax schedule per RA 10963.

````python
# Excise tax per liter by product and TRAIN tranche
# Tranche 1: Jan 2018 | Tranche 2: Jan 2019 | Tranche 3: Jan 2020

def excise_gasoline(date):
    if date < pd.Timestamp('2018-01-01'): return 4.35
    elif date < pd.Timestamp('2019-01-01'): return 7.00
    elif date < pd.Timestamp('2020-01-01'): return 9.00
    else: return 10.00

def excise_diesel(date):
    if date < pd.Timestamp('2018-01-01'): return 0.00
    elif date < pd.Timestamp('2019-01-01'): return 2.65
    elif date < pd.Timestamp('2020-01-01'): return 4.50
    else: return 6.00

def excise_kerosene(date):
    if date < pd.Timestamp('2018-01-01'): return 0.00
    elif date < pd.Timestamp('2019-01-01'): return 1.00
    elif date < pd.Timestamp('2020-01-01'): return 3.00
    else: return 5.00

df['excise_gasoline'] = [excise_gasoline(d) for d in df.index]
df['excise_diesel']   = [excise_diesel(d)   for d in df.index]
df['excise_kerosene'] = [excise_kerosene(d) for d in df.index]

print('Excise tax values by year (diesel):')
print(df.groupby(df.index.year)['excise_diesel'].first())
print('\nExcise tax values by year (gasoline):')
print(df.groupby(df.index.year)['excise_gasoline'].first())
````

## Step 6 — Lag Features

**Lag selection rationale from EDA:**
- CCF MOPS→diesel: lag 1 = 0.417, lag 2 = 0.178, lag 3 = 0.084, lag 4 = 0.021, noise after that
- Own-price autocorrelation ≈ 0.97 — own-price lag 1 is the strongest predictor
- Build lags 1, 2, 3, 4 on all continuous features
- Lag 0 is strictly forbidden on MOPS and crude features (leakage — same-week values not available at announcement time)

**Important:** Lag features are built BEFORE the train/test split. The split happens in the modeling notebook. But we only use lag features starting from lag 1 to prevent leakage.

````python
# Features to lag — excludes targets and dummies
LAG_FEATURES = [
    'mops_gasoil_mean', 'mops_rbob_mean',
    'brent_mean', 'dubai_close',
    'usd_php_mean', 'cpi_value', 'rub_usd_mean',
    'mops_in_php', 'crack_spread',
    'mops_chg_pos', 'mops_chg_neg',
]

# Own-price lags — each product lagged on itself
OWN_PRICE_TARGETS = ['diesel', 'diesel_plus', 'kerosene', 'ron100', 'ron91', 'ron95', 'ron97']

LAGS = [1, 2, 3, 4]

lag_cols_created = []

for col in LAG_FEATURES:
    for lag in LAGS:
        new_col = f'{col}_lag{lag}'
        df[new_col] = df[col].shift(lag)
        lag_cols_created.append(new_col)

for col in OWN_PRICE_TARGETS:
    for lag in LAGS:
        new_col = f'{col}_lag{lag}'
        df[new_col] = df[col].shift(lag)
        lag_cols_created.append(new_col)

print(f'Lag columns created: {len(lag_cols_created)}')
print(f'Total columns now:   {df.shape[1]}')

# Drop rows with NaN from lag construction
rows_before = len(df)
df = df.dropna(subset=[f'mops_gasoil_mean_lag4'])  # lag 4 is the deepest — if it exists, all shallower lags do too
rows_after = len(df)
print(f'\nRows dropped from lag NaN construction: {rows_before - rows_after}')
print(f'Remaining rows: {rows_after}')
print(f'Date range after drop: {df.index.min()} → {df.index.max()}')
````

## Step 7 — p-value Feature Selection (Following Lunor et al.)

Lunor et al. (2023) ran OLS regression of each feature on the target and dropped features with p-value > 0.05. They found that diesel and diesel_plus had different significant feature sets than the RON products. We replicate this per product.

**Scope:** We apply p-value testing to the lag-1 versions of the base features only — these are the primary candidates. Lag 2–4 will be handled by PCA in Step 8. Own-price lags are kept unconditionally (autocorrelation of 0.97 guarantees significance).

**Note:** p-value testing is done on the FULL sample here for feature selection purposes, separate from model training. This is the same approach Lunor et al. used. The actual train/test split happens in the modeling notebook.

````python
from sklearn.linear_model import LinearRegression
from scipy.stats import t as t_dist

# Candidate lag-1 features for p-value testing
CANDIDATE_FEATURES = [f'{c}_lag1' for c in LAG_FEATURES]

def pvalue_feature_selection(df, target, candidates, alpha=0.05):
    """OLS p-value test per feature vs target. Returns significant features."""
    data = df[[target] + candidates].dropna()
    y = data[target].values
    results = []
    
    for feat in candidates:
        X = data[[feat]].values
        # OLS manually to get p-values
        X_with_const = np.column_stack([np.ones(len(X)), X])
        beta = np.linalg.lstsq(X_with_const, y, rcond=None)[0]
        y_hat = X_with_const @ beta
        residuals = y - y_hat
        n, k = X_with_const.shape
        s2 = (residuals @ residuals) / (n - k)
        var_beta = s2 * np.linalg.inv(X_with_const.T @ X_with_const)
        se = np.sqrt(np.diag(var_beta))
        t_stat = beta / se
        p_val = 2 * (1 - t_dist.cdf(np.abs(t_stat), df=n-k))
        results.append({
            'feature': feat,
            'beta': round(beta[1], 4),
            'p_value': round(p_val[1], 4),
            'significant': p_val[1] < alpha
        })
    
    return pd.DataFrame(results).sort_values('p_value')

# Run per product
PRODUCTS = ['diesel', 'ron91', 'ron95', 'ron97']
significant_features_per_product = {}

for product in PRODUCTS:
    result_df = pvalue_feature_selection(df, product, CANDIDATE_FEATURES)
    sig = result_df[result_df['significant']]['feature'].tolist()
    significant_features_per_product[product] = sig
    print(f'\n=== {product.upper()} — significant lag-1 features (p < 0.05) ===')
    print(result_df[['feature','beta','p_value','significant']].to_string(index=False))

print('\n=== SUMMARY: significant features per product ===')
for prod, feats in significant_features_per_product.items():
    print(f'{prod}: {feats}')
````


## Step 8 — PCA on Lag Feature Block

Following Lunor et al. (2023) Section 2.2: after p-value selection, apply PCA to the remaining features to reduce multicollinearity. Retain components explaining 95% of cumulative variance.

**Why PCA is needed here:** From EDA, the crude oil features (MOPS, Brent, Dubai) are 0.90–0.96 correlated with each other. Including all of them plus 4 lags each creates ~12 nearly-identical columns. PCA compresses these into 2-3 orthogonal components.

**Critical:** PCA is fit ONLY on training data to prevent data leakage. Here we fit on the full sample for the purpose of documenting variance explained. In the modeling notebook, PCA will be re-fit inside the walk-forward cross-validation loop on training folds only.

````python
# PCA on the crude oil lag block — the most collinear set
CRUDE_LAG_COLS = [
    f'{feat}_lag{lag}'
    for feat in ['mops_gasoil_mean', 'brent_mean', 'dubai_close', 'mops_in_php']
    for lag in LAGS
]
crude_lag_cols = [c for c in CRUDE_LAG_COLS if c in df.columns]

pca_data = df[crude_lag_cols].dropna()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(pca_data)

pca = PCA()
pca.fit(X_scaled)

cumvar = np.cumsum(pca.explained_variance_ratio_)
n_components_95 = int(np.argmax(cumvar >= 0.95) + 1)

print(f'Crude oil lag block: {len(crude_lag_cols)} columns')
print(f'\nVariance explained per component:')
for i, (ev, cv) in enumerate(zip(pca.explained_variance_ratio_, cumvar)):
    marker = ' ← 95% threshold' if i+1 == n_components_95 else ''
    print(f'  PC{i+1}: {ev:.4f} individual, {cv:.4f} cumulative{marker}')
    if cv > 0.99: break

print(f'\nComponents needed for 95% variance: {n_components_95}')
print(f'Compression: {len(crude_lag_cols)} columns → {n_components_95} components')
print('\nNote: PCA will be RE-FIT inside walk-forward CV folds in the modeling notebook.')
print('This fit is for documentation purposes only.')
````

## Step 9 — Build Final Feature Matrices Per Product

Following Lunor et al.'s approach of per-product feature sets. Each product gets:
- Its own-price lags (unconditionally included — autocorr ≈ 0.97)
- The p-value significant lag-1 features for that product
- Lags 2–4 of MOPS and Brent (for PCA)
- Asymmetric MOPS features (RFH, lags 1–2)
- Event dummies and excise tax
- The corresponding target column

````python
# Always-included features regardless of p-value
ALWAYS_INCLUDE = [
    # Event dummies
    'event_covid', 'event_ru_ua', 'event_tariff',
    # Asymmetric MOPS (RFH features)
    'mops_chg_pos_lag1', 'mops_chg_neg_lag1',
    'mops_chg_pos_lag2', 'mops_chg_neg_lag2',
]

# Crude lag block for PCA (lags 2-4 of MOPS, Brent, Dubai)
CRUDE_FOR_PCA = [
    f'{feat}_lag{lag}'
    for feat in ['mops_gasoil_mean', 'brent_mean', 'dubai_close']
    for lag in [2, 3, 4]
]
crude_for_pca = [c for c in CRUDE_FOR_PCA if c in df.columns]

product_feature_sets = {}
for product in PRODUCTS:
    excise_col = 'excise_diesel' if product in ['diesel','diesel_plus','kerosene'] else 'excise_gasoline'
    own_lags = [f'{product}_lag{lag}' for lag in LAGS]
    
    feature_set = (
        own_lags
        + significant_features_per_product.get(product, [])
        + ALWAYS_INCLUDE
        + crude_for_pca
        + [excise_col]
    )
    # Deduplicate while preserving order
    seen = set()
    feature_set = [f for f in feature_set if f in df.columns and not (f in seen or seen.add(f))]
    product_feature_sets[product] = feature_set
    print(f'{product}: {len(feature_set)} features')

print('\nDiesel feature set:')
print(product_feature_sets['diesel'])
````

## Step 10 — Save Feature Matrix

Save the full enriched dataframe (all columns, all lags) plus the per-product feature set definitions. The modeling notebook will load both and handle train/test splitting and PCA fitting inside folds.

````python
import json, os
os.makedirs('../data/final', exist_ok=True)

# Save full enriched dataframe
df.to_parquet('../data/final/features_engineered.parquet')
print('Saved features_engineered.parquet')
print('Shape:', df.shape)
print('Date range:', df.index.min(), '→', df.index.max())

# Save per-product feature sets as JSON
with open('../data/final/feature_sets.json', 'w') as f:
    json.dump(product_feature_sets, f, indent=2)
print('\nSaved feature_sets.json')

# Final summary
print('\n=== FEATURE ENGINEERING DECISION LOG ===')
print(f'WTI:          DROPPED (Brent-WTI r=0.991)')
print(f'Nulls:        Forward-filled (3 diesel/kerosene, 27 ron100)')
print(f'Lags built:   1-4 on {len(LAG_FEATURES)} features + {len(OWN_PRICE_TARGETS)} own-price series')
print(f'Derived:      mops_in_php, crack_spread, mops_chg_pos, mops_chg_neg')
print(f'Events:       3 binary dummies (COVID, Russia-Ukraine, tariff)')
print(f'Excise:       3 step-function columns from TRAIN Law schedule')
print(f'p-value sel:  Per product, alpha=0.05, lag-1 features only')
print(f'PCA:          Will be applied inside CV folds in modeling notebook')
print(f'Final rows:   {len(df)} (after dropping first 4 rows for lag construction)')
print(f'Ready for:    04_Modeling.ipynb')
````
