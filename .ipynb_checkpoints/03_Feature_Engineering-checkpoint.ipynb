# Notebook 03: Feature Engineering
**Philippine NCR Pump Price Forecasting**

This notebook constructs all model-ready features from the master weekly table.
Every feature is grounded in a specific research justification:

| Feature Group | Research Basis |
|---|---|
| `mops_in_php` (MOPS × FX) | DOE pricing formula (direct mechanism) |
| Lag features (lag 1–4) | Ljubić et al. (2023) shifting experiment; DOE 1-week lag |
| Rolling mean/std | Captures volatility regime; Lunor et al. (2023) |
| Asymmetric returns r+, r- | Wen et al. (2025) Rockets-and-Feathers Hypothesis |
| Month dummies + trend | He (2023) LM1 specification |
| Event dummies | Lunor et al. (2023); structural breaks identified in EDA |
| Excise tax step | TRAIN Law RA 10963; directly in DOE formula |

### Output
`data/final/feature_matrix.parquet` — one row per week, all features, ready for modelling

````python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

os.makedirs('../data/final', exist_ok=True)
os.makedirs('../outputs', exist_ok=True)

print('Libraries loaded.')
````

````python
master = pd.read_parquet('data/final/master_weekly.parquet')

# Products we will model
PRODUCTS = ['diesel', 'ron91', 'ron95', 'ron97']
available_products = [p for p in PRODUCTS if p in master.columns]

# Drop WTI — redundant with Brent (>0.95 corr, US-centric)
# Decision justified in EDA notebook Section 3
if 'wti_mean' in master.columns:
    master = master.drop(columns=[c for c in master.columns if c.startswith('wti')])
    print('Dropped WTI columns (redundant with Brent, justified in EDA §3)')

print('Master shape:', master.shape)
print('Date range:', master.index.min(), '->', master.index.max())
master.head(3)
````

## Feature Group 1: Formula-Derived Feature

**Research basis**: The DOE pricing formula (§0 of methodology doc) is:
```
PumpPrice = MOPS × USD_PHP + Freight + Excise + VAT + Margin
```
The product `MOPS × USD_PHP` is the single most important input. He (2023) showed crude oil alone (R²=0.874) is far worse than crude + macro features (R²=0.947). For PH, `mops_in_php` captures both the commodity and FX component in one term — the dominant predictor.

````python
df = master.copy()

# mops_in_php: directly encodes the PH pricing formula
# MOPS_gasoil × USD/PHP = cost of refined product in pesos before taxes
if 'mops_gasoil_mean' in df.columns and 'usd_php_mean' in df.columns:
    df['mops_in_php_gasoil'] = df['mops_gasoil_mean'] * df['usd_php_mean']
    print('Created: mops_in_php_gasoil (MOPS gasoil × USD/PHP)')

if 'mops_rbob_mean' in df.columns and 'usd_php_mean' in df.columns:
    df['mops_in_php_gasoline'] = df['mops_rbob_mean'] * df['usd_php_mean']
    print('Created: mops_in_php_gasoline (MOPS RBOB × USD/PHP)')

# Also compute crack spread: MOPS_gasoil - Brent (refining margin)
# Asche et al. (2003): this isolates refining cost dynamics from crude
if 'mops_gasoil_mean' in df.columns and 'brent_mean' in df.columns:
    df['crack_spread_gasoil'] = df['mops_gasoil_mean'] - df['brent_mean']
    print('Created: crack_spread_gasoil (MOPS gasoil - Brent, refining margin proxy)')

print()
print('These are the mechanistically grounded features — they should dominate SHAP plots later.')
````

## Feature Group 2: Lag Features

**Research basis**: Ljubić et al. (2023) §4.3 found that Brent leads Bosnian fuel prices by ~32 days. Our EDA cross-correlation experiment (Notebook 02 §8) quantifies the exact lag for PH data. The DOE mechanism guarantees lag 1 is already built in by our `W-FRI + 3 days` aggregation. Lags 2–4 capture slower pass-through dynamics.

````python
# Features to lag — these are the key exogenous inputs
lag_base_features = [
    'brent_mean', 'usd_php_mean',
    'mops_gasoil_mean', 'mops_in_php_gasoil',
]
lag_base_features = [f for f in lag_base_features if f in df.columns]

# Build lags 2, 3, 4 (lag 1 is already encoded by our W-FRI aggregation)
for col in lag_base_features:
    for lag in [2, 3, 4]:
        df[f'{col}_lag{lag}'] = df[col].shift(lag)

# Autoregressive lag of pump price itself
# Important: lag 1 of the target is a powerful predictor but note:
# Eliwa et al. (2024) warned that including prev_price alone inflates R² to ~0.99
# by essentially outputting persistence. We include it but evaluate with MASE.
for product in available_products:
    df[f'{product}_lag1'] = df[product].shift(1)
    df[f'{product}_lag2'] = df[product].shift(2)

print('Lag features created:')
lag_cols = [c for c in df.columns if '_lag' in c]
print(f'  Total lag columns: {len(lag_cols)}')
print(f'  Columns: {lag_cols[:10]} ...')
````

## Feature Group 3: Rolling Statistics

**Research basis**: Rolling mean and std capture the price regime and volatility state. This is particularly important for the Rockets-and-Feathers asymmetry — Wen et al. (2025) show that economic conditions moderate the RFH effect. A 4-week rolling std captures whether we are in a high-volatility regime.

````python
roll_features = ['brent_mean', 'usd_php_mean']
roll_features = [f for f in roll_features if f in df.columns]

for col in roll_features:
    df[f'{col}_roll4_mean'] = df[col].rolling(4).mean()
    df[f'{col}_roll4_std']  = df[col].rolling(4).std()
    df[f'{col}_roll8_mean'] = df[col].rolling(8).mean()

# Brent realized volatility (rolling 4-week std of returns)
if 'brent_mean' in df.columns:
    brent_ret = df['brent_mean'].pct_change()
    df['brent_realized_vol'] = brent_ret.rolling(4).std()
    print('Created: brent_realized_vol (4-week rolling std of Brent returns)')

print('Rolling features created:')
roll_cols = [c for c in df.columns if 'roll' in c or 'realized_vol' in c]
print(f'  Total: {len(roll_cols)} columns')
````

## Feature Group 4: Asymmetric Returns (Wen et al. 2025 — RFH)

**Research basis**: Wen et al. (2025) Eq. 2 and 3 — split crude oil weekly returns into positive and negative components:
- `r+ = max(r, 0)`: upward moves only
- `r- = min(r, 0)`: downward moves only

Their finding: **only r+ is statistically significant** (β = 0.178–0.215, p<0.01). r- is insignificant. This is the Rockets-and-Feathers effect: price hikes pass through immediately, rollbacks do not.

This is our **novelty hook** — we test whether this asymmetry holds for PH pump prices, which has not been published before.

````python
if 'brent_mean' in df.columns:
    df['brent_return']  = df['brent_mean'].pct_change()
    df['brent_ret_pos'] = df['brent_return'].clip(lower=0)   # r+ = max(r, 0)
    df['brent_ret_neg'] = df['brent_return'].clip(upper=0)   # r- = min(r, 0)

    # Lagged asymmetric returns (for multi-step forecasting)
    df['brent_ret_pos_lag1'] = df['brent_ret_pos'].shift(1)
    df['brent_ret_neg_lag1'] = df['brent_ret_neg'].shift(1)

    print('Created asymmetric return features (Wen et al. 2025):')
    print('  brent_return     : raw weekly pct change')
    print('  brent_ret_pos    : r+ = max(r, 0), upward moves only')
    print('  brent_ret_neg    : r- = min(r, 0), downward moves only')
    print()

    # Quick visualization: do positive and negative returns behave differently?
    fig, axes = plt.subplots(1, 3, figsize=(14, 3))
    df['brent_return'].dropna().hist(bins=50, ax=axes[0], color='gray')
    axes[0].set_title('r (raw returns)')
    df['brent_ret_pos'].dropna().hist(bins=50, ax=axes[1], color='tomato')
    axes[1].set_title('r+ (positive only)')
    df['brent_ret_neg'].dropna().hist(bins=50, ax=axes[2], color='steelblue')
    axes[2].set_title('r- (negative only)')
    for ax in axes:
        ax.grid(True, alpha=0.3)
    plt.suptitle('Brent Return Decomposition (Wen et al. 2025 RFH features)', fontsize=11)
    plt.tight_layout()
    plt.savefig('../outputs/03a_asymmetric_returns.png', dpi=150, bbox_inches='tight')
    plt.show()
````

## Feature Group 5: Calendar Features

**Research basis**: He (2023) LM1 model uses trend variable `t` and 11 monthly seasonal dummies (S2–S12) — this is his best-performing specification (R²=0.947 vs 0.874 for crude-only). We replicate this.

````python
# Time trend (sequential integer, week 1, 2, 3...)
df['trend_t'] = np.arange(len(df))

# Month of year (1–12)
df['month'] = df.index.month

# Month dummy variables (drop month 1 as reference — He 2023 approach)
month_dummies = pd.get_dummies(df['month'], prefix='month', drop_first=True)
df = pd.concat([df, month_dummies], axis=1)

# ISO week of year
df['iso_week'] = df.index.isocalendar().week.astype(int)

print('Calendar features created:')
print('  trend_t     : sequential week counter (1, 2, 3, ...)')
print('  month       : month of year (1-12)')
print('  month_2 to month_12: dummy variables (month_1 = reference)')
print('  iso_week    : ISO week of year')
print(f'  Total month dummies: {len([c for c in df.columns if c.startswith("month_") and c != "month_"])}')
````

## Feature Group 6: Event Dummy Variables

**Research basis**: Lunor et al. (2023) explicitly modelled COVID-19 and Russia-Ukraine conflict as binary features. Our methodology doc §3.3 specifies the exact date windows.

````python
# COVID lockdown period
df['event_covid'] = (
    (df.index >= '2020-03-16') & (df.index <= '2020-09-30')
).astype(int)

# Russia-Ukraine war
df['event_russia_ukraine'] = (
    df.index >= '2022-02-24'
).astype(int)

print('Event dummy variables created:')
print(f'  event_covid          : {df["event_covid"].sum()} weeks (Mar–Sep 2020)')
print(f'  event_russia_ukraine : {df["event_russia_ukraine"].sum()} weeks (Feb 2022 onward)')
````

## Feature Group 7: Excise Tax Step Variable

**Research basis**: TRAIN Law (RA 10963) raised excise taxes in three tranches. These are step functions, not continuous. Without encoding them, a model will mistake the step changes in price as noise. This is a critical feature for avoiding the structural-break problem (methodology doc §4.3).

````python
# Diesel excise tax (₱/L) per TRAIN law schedule
# Pre-TRAIN: 0, Jan 2018: ₱2.65, Jan 2019: ₱4.50, Jan 2020: ₱6.00
df['excise_diesel'] = 0.0
df.loc[df.index >= '2018-01-01', 'excise_diesel'] = 2.65
df.loc[df.index >= '2019-01-01', 'excise_diesel'] = 4.50
df.loc[df.index >= '2020-01-01', 'excise_diesel'] = 6.00

# Gasoline excise tax (₱/L)
# Pre-TRAIN: 4.35, Jan 2018: ₱7.00, Jan 2019: ₱9.00, Jan 2020: ₱10.00
df['excise_gasoline'] = 0.0
df.loc[df.index >= '2018-01-01', 'excise_gasoline'] = 7.00
df.loc[df.index >= '2019-01-01', 'excise_gasoline'] = 9.00
df.loc[df.index >= '2020-01-01', 'excise_gasoline'] = 10.00

print('Excise tax step variables created (TRAIN Law RA 10963):')
print('  excise_diesel   — diesel excise ₱/L')
print('  excise_gasoline — gasoline excise ₱/L')
print()
print('Diesel excise distribution:')
print(df['excise_diesel'].value_counts().sort_index())
````

## Finalize Feature Matrix

Drop NaN rows from lag construction, drop WTI, clip to 2018 onward.

````python
# Trim to 2018 onward (consistent excise structure post-TRAIN tranche 1)
df = df.loc['2018-01-01':].copy()

# Drop rows with NaN from lag construction (first ~4 rows)
df = df.dropna(subset=[c for c in df.columns if '_lag' in c or 'roll' in c], how='all')

# Show feature summary
feature_cols = [c for c in df.columns if c not in available_products]
target_cols  = available_products

print(f'Feature matrix shape: {df.shape}')
print(f'Feature columns: {len(feature_cols)}')
print(f'Target columns:  {target_cols}')
print(f'Date range: {df.index.min()} to {df.index.max()}')
print(f'Remaining nulls (features):')
null_check = df[feature_cols].isnull().sum()
print(null_check[null_check > 0])
````

````python
# Save feature matrix
df.to_parquet('../data/final/feature_matrix.parquet')
print('Saved: data/final/feature_matrix.parquet')
print()
print('All feature columns:')
for c in sorted(feature_cols):
    print(f'  {c}')
````

## Feature Correlation Heatmap (post-engineering)

````python
import seaborn as sns

# Focus on key non-lag features vs diesel for clarity
key_features = [
    'mops_in_php_gasoil', 'brent_mean', 'usd_php_mean',
    'mops_gasoil_mean', 'crack_spread_gasoil',
    'brent_ret_pos', 'brent_ret_neg',
    'brent_realized_vol', 'cpi_value',
    'excise_diesel', 'event_covid', 'event_russia_ukraine',
    'trend_t'
]
key_features = [f for f in key_features if f in df.columns]
target = [p for p in available_products if p in df.columns]

corr = df[key_features + target].dropna().corr()

fig, ax = plt.subplots(figsize=(13, 10))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='RdYlGn',
            vmin=-1, vmax=1, ax=ax, annot_kws={'size': 7},
            linewidths=0.3)
ax.set_title('Feature Correlation Matrix (Post Engineering)', fontsize=12, weight='bold')
plt.tight_layout()
plt.savefig('../outputs/03b_feature_correlation.png', dpi=150, bbox_inches='tight')
plt.show()

if 'diesel' in corr.columns:
    print('\nCorrelation with diesel (sorted):')
    print(corr['diesel'].drop('diesel').sort_values(ascending=False).round(3))
````
